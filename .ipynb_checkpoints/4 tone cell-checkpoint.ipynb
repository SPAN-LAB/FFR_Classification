```python
import os
import numpy as np, h5py, tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.model_selection import StratifiedKFold, StratifiedShuffleSplit
from sklearn.metrics import accuracy_score, confusion_matrix
import pandas as pd

BASE_DIR = "/Volumes/gurindapalli/projects/trial_classification/4tone_cell"
OUT_DIR  = "/Volumes/gurindapalli/projects/trial_classification/Jason"
os.makedirs(OUT_DIR, exist_ok=True)

DS_FACTOR   = 8        
EPOCHS      = 15
BATCH_SIZE  = 32
N_SPLITS    = 5
VAL_SPLIT   = 0.10
N_CLASSES   = 4
```

```python
def build_model(input_len, n_classes=4, l2=1e-5, sdrop=0.10, head_drop=0.30):
    inp = layers.Input(shape=(input_len,1))
    x = layers.Conv1D(128, 9, padding="same", activation="relu")(inp)
    x = layers.SpatialDropout1D(sdrop)(x)
    x = layers.MaxPooling1D(2)(x)

    x = layers.Conv1D(128, 9, padding="same", activation="relu")(x)
    x = layers.SpatialDropout1D(sdrop)(x)
    x = layers.MaxPooling1D(2)(x)

    x = layers.Conv1D(256, 9, padding="same", activation="relu")(x)
    x = layers.SpatialDropout1D(sdrop)(x)

    x = layers.GlobalAveragePooling1D()(x)
    x = layers.Dropout(head_drop)(x)

    out = layers.Dense(
        n_classes, activation="softmax",
        kernel_regularizer=tf.keras.regularizers.l2(l2)
    )(x)

    model = models.Model(inp, out)
    model.compile(
        optimizer=tf.keras.optimizers.Adam(1e-3),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model
```

```python
def subaverage(X, y, n_classes=N_CLASSES, group_size=5, repeats=6, bootstrap=True, renorm=True):
    rng = np.random.default_rng()
    X_new, y_new = [], []
    for c in range(n_classes):
        idx = np.where(y == c)[0]
        if len(idx) < group_size:
            continue
        n_groups = max(1, (len(idx) // group_size) * repeats)
        for _ in range(n_groups):
            take = rng.choice(idx, size=group_size, replace=bootstrap)
            x_avg = X[take].mean(axis=0)
            X_new.append(x_avg); y_new.append(c)
    if not X_new:
        return None, None
    return np.stack(X_new, axis=0), np.array(y_new, dtype=np.int32)
```

```python
def load_4tone_safe(mat_path, dataset="ffr_dss", ds_factor=DS_FACTOR):
    with h5py.File(mat_path, "r", locking=False) as f:
        X = f[dataset][()]
        if   X.shape[0] == 4997: X = X.T
        elif X.shape[1] == 4997: pass
        else:
            raise ValueError(f"{dataset} shape {X.shape}: no time=4997 axis")

        N = X.shape[0]
        y = None

        if "labels" in f:
            lab = f["labels"]
            if lab.ndim == 2 and lab.shape[1] == N and ("object" in str(lab.dtype) or "O" in str(lab.dtype)):
                y = np.array([int(np.array(f[lab[0, i]])[0, 0]) for i in range(N)], dtype=int)
            elif lab.ndim == 2 and lab.shape[1] == N and np.issubdtype(lab.dtype, np.integer):
                y = lab[()].ravel().astype(int)

        # (B) 4T1005/1014-style "#refs#/h"
        if (y is None or y.size != N) and "#refs#" in f and "h" in f["#refs#"]:
            h = np.atleast_1d(f["#refs#/h"][()]).ravel().astype(int)
            if h.size == N:
                y = h
    u = set(np.unique(y).tolist())
    if u == {49,50,51,52}:
        y = (y - 49).astype(np.int32)    
    elif u == {1,2,3,4}:
        y = (y - 1).astype(np.int32)    
    else:
        tone_to_idx = {v:i for i,v in enumerate(sorted(u))}
        y = np.array([tone_to_idx[v] for v in y], dtype=np.int32)

    if ds_factor:
        X = X[:, ::ds_factor]
    m = X.mean(axis=1, keepdims=True); s = X.std(axis=1, keepdims=True) + 1e-7
    X = ((X - m) / s)[..., np.newaxis]
    return X, y

def counts(v): 
    return np.bincount(v, minlength=N_CLASSES).tolist()
```

```python
def run_cv_on_file(mat_path):
    X, y = load_4tone_safe(mat_path, dataset="ffr_dss", ds_factor=DS_FACTOR)
    base = os.path.basename(mat_path)
    print(f"{base} -> X={X.shape}, y={y.shape}, uniques={np.unique(y)}")
    print("class counts:", np.bincount(y, minlength=N_CLASSES).tolist())

    skf = StratifiedKFold(n_splits=N_SPLITS, shuffle=True)
    fold_accs = []
    cm_total  = np.zeros((N_CLASSES, N_CLASSES), dtype=int)

    for i, (tr, te) in enumerate(skf.split(X, y), 1):
        sss = StratifiedShuffleSplit(n_splits=1, test_size=VAL_SPLIT)
        tr_sub_idx, val_idx = next(sss.split(X[tr], y[tr]))
        Xtr, ytr = X[tr][tr_sub_idx], y[tr][tr_sub_idx]
        Xval, yval = X[tr][val_idx],   y[tr][val_idx]
        Xte,  yte  = X[te],            y[te]

        Xtr_sa, ytr_sa = subaverage(Xtr, ytr, n_classes=N_CLASSES, group_size=5, repeats=6, bootstrap=True, renorm=True)
        Xtr_parts, ytr_parts = [Xtr], [ytr]
        if Xtr_sa is not None:
            Xtr_parts.append(Xtr_sa); ytr_parts.append(ytr_sa)
        Xtr_aug = np.concatenate(Xtr_parts, axis=0)
        ytr_aug = np.concatenate(ytr_parts, axis=0)

        model = build_model(X.shape[1], N_CLASSES)

        priors = np.bincount(ytr_aug, minlength=N_CLASSES) / len(ytr_aug)
        model.layers[-1].bias.assign(np.log(priors + 1e-8))

        class_counts = np.bincount(ytr_aug, minlength=N_CLASSES)
        class_weight = dict(enumerate((len(ytr_aug) / (N_CLASSES * np.maximum(class_counts,1))).astype(np.float32)))

        es  = tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=4, restore_best_weights=True)
        rlr = tf.keras.callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=2, min_lr=1e-5)

        print(f"\n{base} | Fold {i}: counts  "
              f"train={counts(ytr)}  "
              f"val={counts(yval)}  "
              f"test={counts(yte)}  "
              f"priors={np.round(priors,3).tolist()}")

        model.fit(
            Xtr_aug, ytr_aug,
            validation_data=(Xval, yval),
            epochs=EPOCHS, batch_size=BATCH_SIZE, verbose=0,
            callbacks=[es, rlr],
            class_weight=class_weight
        )

        y_hat = np.argmax(model.predict(Xte, batch_size=BATCH_SIZE, verbose=0), axis=1)
        acc   = accuracy_score(yte, y_hat)
        fold_accs.append(acc)
        cm_total += confusion_matrix(yte, y_hat, labels=list(range(N_CLASSES)))

        pred_counts = np.bincount(y_hat, minlength=N_CLASSES)
        print(f"{base} | Fold {i}: acc={acc:.4f} | pred counts={pred_counts.tolist()}")

    print(f"{base} | Mean acc: {np.mean(fold_accs):.4f} ± {np.std(fold_accs):.4f}")
    return fold_accs, cm_total
```

```python
file_ids   = list(range(1002, 1016)) 
accs_by_file = {}
labels       = []
agg_rows     = []

for fid in file_ids:
    fname = f"4T{fid}.mat"
    mat_path = os.path.join(BASE_DIR, fname)
    if not os.path.exists(mat_path):
        print(f"Missing file: {mat_path} (skipping)")
        continue

    fold_accs, cm_total = run_cv_on_file(mat_path)
    key = f"4T{fid}"
    accs_by_file[key] = fold_accs
    labels.append(key)

    row = {"file": key}
    for i in range(N_CLASSES):
        for j in range(N_CLASSES):
            row[f"cm_{i}{j}"] = int(cm_total[i, j])
    agg_rows.append(row)

    tf.keras.backend.clear_session()

agg_cols = ["file"] + [f"cm_{i}{j}" for i in range(N_CLASSES) for j in range(N_CLASSES)]
df_agg = pd.DataFrame(agg_rows, columns=agg_cols)
agg_csv_path = os.path.join(OUT_DIR, "cm_all_files1.csv")
df_agg.to_csv(agg_csv_path, index=False)
print(f"\nSaved aggregated confusion matrices → {agg_csv_path}")

print("\nPer-file mean ± SD across CV folds:")
for k in labels:
    m  = float(np.mean(accs_by_file[k]))
    sd = float(np.std(accs_by_file[k]))
    print(f"{k}: {m:.4f} ± {sd:.4f}")
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import display

csv = "/Volumes/gurindapalli/projects/trial_classification/Jason/cm_all_files1.csv"

agg = pd.read_csv(csv)
required_cols = ["file"] + [f"cm_{i}{j}" for i in range(4) for j in range(4)]
missing = [c for c in required_cols if c not in agg.columns]
if missing:
    raise ValueError(f"CSV missing expected columns: {missing}")

tone_labels = [1, 2, 3, 4]

def row_to_cm(row):
    cm = np.zeros((4,4), dtype=int)
    for i in range(4):
        for j in range(4):
            cm[i, j] = int(row[f"cm_{i}{j}"])
    return cm
records = []  
for _, r in agg.iterrows():
    tag = str(r["file"])
    cm  = row_to_cm(r)

    df_cm = pd.DataFrame(cm,
                         index=[f"true_{t}" for t in tone_labels],
                         columns=[f"pred_{t}" for t in tone_labels])
    print(f"\n=== {tag}: Confusion Matrix (rows=true, cols=pred) ===")
    display(df_cm.astype(int).style.format("{:d}"))

    row_tot = cm.sum(axis=1)
    with np.errstate(divide="ignore", invalid="ignore"):
        recalls = np.diag(cm) / np.where(row_tot == 0, np.nan, row_tot)

    for t, val in enumerate(recalls, start=1):
        if not np.isnan(val):
            records.append({"file": tag, "tone": int(t), "recall": float(val)})

df_long = pd.DataFrame(records)

summary = (df_long.groupby("tone")["recall"]
           .agg(n_files="count", mean="mean", sd="std", median="median")
           .reset_index())
print("\n--- Per-tone summary across files ---")
display(summary.style.format({"mean": "{:.3f}", "sd": "{:.3f}", "median": "{:.3f}"}))

sns.set(style="whitegrid", context="talk")
plt.figure(figsize=(9, 5.5))
ax = sns.boxplot(data=df_long, x="tone", y="recall")
ax.set_xlabel("Tone")
ax.set_ylabel("Per-tone correct rate (recall)")
ax.set_title("Per-tone performance across files (from aggregated confusion matrices CSV)")
ax.set_ylim(0, 1)
plt.tight_layout()
plt.show()
```
